```python
def apply_keywords(text):
    # 딕셔너리의 각 value 순회 - 해당하는 경우 그 단어는 key값으로 변환하기
    key_list = okt.nouns(text)
    for k, v in synonyms.items():
        for i, key in enumerate(key_list):
            if key in v:
                key_list[i] = k
    return key_list
```

## 유사도 계산

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_by_preference(preference_vector, tfidf_matrix, top_n=10):
    # 코사인 유사도 계산
    sim_scores = cosine_similarity(preference_vector, tfidf_matrix).flatten()
    
    # 유사도 점수를 기준으로 정렬하여 가장 유사한 문서의 인덱스 찾기
    top_indices = np.argsort(sim_scores)[::-1][:top_n]
    
    return top_indices
```

```python
recommended_document = recommend_by_preference(user_vector, matrix, 15)
print("Recommended Document:", recommended_document)
```
