## 유사한 단어들은 키워드로 치환
- GPT 사용하여 500개의 키워드 분류

```python
keywords = ["산맥", "바다", "동굴", "체험", "유적지", "지질시대", "산책로", "계곡", "하천"]

synonyms = {"산맥" : ["산", "운문", "트레일", "출발", "대포", "화산", "암석", "지질", "용암", "현무암", "유문암", "산책로", "주흘산", "트레킹", "산책"]
,
"바다" : ["해안가", "해안", "절경", "해수욕장", "송도", "해상", "갯벌", "연안", "철새", "해양"],

"체험" : ["체험", "자연", "생태관광", "관광", "생태도시", "친환경", "산책로", "트레킹", "관광객", "인식", "교육장", "생활"],

"유적지" : ["조선", "숙종", "인조", "영남", "문경", "관광명소"],

"지질시대" : ["백악기", "지질", "단층", "암맥", "관입", "연흔", "석회질", "토양", "지각", "변형", "관문", "퇴적암", "화쇄류", "공룡", "화석"],

"산책로" : ["산책로", "산책", "트레일", "둘레길", "완만", "휴식", "맨발", "자연", "관광"],

"계곡" : ["계곡", "폭포", "용추", "천지연폭포", "상류", "주흘산", "귀정"],

"하천" : ["강", "하천", "계곡", "하부", "한강"]}
```

```python
def apply_keywords(text):
    # 딕셔너리의 각 value 순회 - 해당하는 경우 그 단어는 key값으로 변환하기
    key_list = okt.nouns(text)
    for k, v in synonyms.items():
        for i, key in enumerate(key_list):
            if key in v:
                key_list[i] = k
    return key_list
```

```python
sample['summaryList'] = sample['cleanSummary'].apply(lambda x : apply_keywords(x))
sample
```

## 유사도 계산

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_by_preference(preference_vector, tfidf_matrix, top_n=10):
    # 코사인 유사도 계산
    sim_scores = cosine_similarity(preference_vector, tfidf_matrix).flatten()
    
    # 유사도 점수를 기준으로 정렬하여 가장 유사한 문서의 인덱스 찾기
    top_indices = np.argsort(sim_scores)[::-1][:top_n]
    
    return top_indices
```

```python
recommended_document = recommend_by_preference(user_vector, matrix, 15)
print("Recommended Document:", recommended_document)
```
